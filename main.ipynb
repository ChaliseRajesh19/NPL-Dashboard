```python
final_matches = matches_sorted.groupby('season ').tail(1)   
final_matches['runner_up'] = final_matches.apply(
    lambda row: row['team_1'] if row['winner'] == row['team_2'] else row['team_2'], axis=1
)
final_matches[['season ', 'winner', 'runner_up']]
```

```python
batsmen= pd.read_csv("NPL/Batting Records/most_runs.csv")
bowlers= pd.read_csv("NPL/Bowling Records/most_wickets.csv")
```

```python
batsmen.shape
```

```python
bowlers.shape
```

```python
# --- STEP 1: Clean player names to avoid mismatch ---
batsmen['player'] = batsmen['player'].astype(str).str.strip().str.lower()
bowlers['player'] = bowlers['player'].astype(str).str.strip().str.lower()

# --- STEP 2: Prefix columns to avoid name conflict ---
batsmen = batsmen.add_prefix("bat_")
bowlers = bowlers.add_prefix("bowl_")

# Restore player column
batsmen = batsmen.rename(columns={"bat_player": "player"})
bowlers = bowlers.rename(columns={"bowl_player": "player"})

# --- STEP 3: Merge both datasets ---
player_df = pd.merge(
    batsmen,
    bowlers,
    on="player",
    how="outer"      # include all players from batting and bowling
)

# --- STEP 4: Replace missing values with ZERO ---
player_df = player_df.fillna(0)

# --- STEP 5: Convert numeric columns ---
for col in player_df.columns:
    if col != "player":
        player_df[col] = pd.to_numeric(player_df[col], errors="ignore")

player_df.head()
```

```python

player_df['player']= player_df['player'].str.replace(r"\(.*\)", "", regex=True).str.strip()
```

```python
player_df.sample(5)
```

```python
player_df.to_csv("NPL/player_dataset.csv", index=False)
player_df.sample(5)
```
